```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,np.pi*2,1000)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y1)
ax.plot(x,y2)

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")

ax.set_title("Karl Roth's Sine and Cosine")
```
